# Sample View

Quick exploration of the Causal-Supreme dataset structure.

```python
import json
import pandas as pd
from collections import Counter
```

## Load P3SA Samples

```python
with open('../data/splits/openai_subsets/p3sa_500.jsonl') as f:
    p3sa = [json.loads(line) for line in f]

print(f"Total P3SA samples: {len(p3sa)}")
```

```python
# View first sample
sample = p3sa[0]
print("Fields:", list(sample.keys()))
print("\nDAG:", sample['dag'])
print("\nQuery:", sample['query'])
print("\nPrompt:", sample['nl_prompt'])
print("\nAnswer:", sample['binary_answer'])
```

## Rung Distribution

```python
rungs = [s['query']['rung'] for s in p3sa]
print("Rung distribution:", Counter(rungs))
```

## Load CLadder Samples

```python
cladder = pd.read_csv('../data/splits/openai_subsets/cladder_600.csv')
print(f"Total CLadder samples: {len(cladder)}")
print("\nColumns:", list(cladder.columns))
print("\nRung distribution:")
print(cladder['rung'].value_counts())
```

```python
# Sample CLadder prompt
print("Sample prompt:")
print(cladder['prompt'].iloc[0][:500])
```
